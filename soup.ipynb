```python
from bs4 import BeautifulSoup
import requests
```

```python
url=("https://www.iplt20.com/")

```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
print(soup)
```

```python
title=soup.find('title')
print(title)
```
